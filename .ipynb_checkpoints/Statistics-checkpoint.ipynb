## Theory

__Statistics:__ It is a branch of mathematics concerns with the collection, organization, analysis, interpretation and collection of data.

__Types of statistics:__
- Descriptive statistics: Describing collected data (Variance, Mean, Mode, Median, Distribution graphs)
- Inferential statistics: Drawing conclusions about population from a sample of population (Hypothesis testing)

__Population:__ Pool of indivisuals from which sample is drawn. In simple words we can say that sample is a subset of population.

__Mean:__ average of all the numeric values

__Median:__ 
- If odd number of values are present in particular distribution then the middle value/entity will be the median.
- If even number of values are present in particular distribution then the average of middle two values will be the median. 

__Mode:__ value/entity with frequent occurance in particular distribution.
 
__Variance:__ measures the spread of the data.

__Covariance:__ measures the direction of correlation between data points of two features.

__Standard deviation:__ is a measure of the amount of variation or dispersion of a set of values, expressing how close the values to the mean of the set/sample.


__Types of data:__
- Numeric data:
  - __`Descrete data`__ (EmpID, Pincode)
  - __`Continuous data`__ (Height, Weight, Salary)
  - __`Categorical data`__ (Gender)

__Gaussian normal distribution:__ is a bell shaped curv having mean equals to 0 and standard deviation as 1. (mean=median=mode)


__Right skewed data:__ also known as positively skewed data where (mean>median>mode)

__Left skewed data:__ also known as negatively skewed data where (mode>median>mean)


__Coeffiecient of correlation:__
- Pearson's coe of coeffiecient (deals with linear data)
- Spearman's coe of coeffiecient (deals with non linear data)


__Five number summary:__ mostly used in outliers detection


__Hypothesis testing:__
- If test value falls in confidence interval we accept null hypothesis and reject alternative hypothesis.
- In simple words, if p_value is greater than significance value(0.05) then we will accept null hypothesis and reject alternative hypothesis.
- If test value does not fall in confidence interval or if test value falls in extreme end then we will reject null hypothesis and accept alternative hypothesis.
- In simple words, if p_value is less than significance value(0.05) then we will reject null hypothesis and accept alternative hypothesis. This means that the test value does not fall in confidence interval, it falls in extreme end.

__Hypothesis:__ is nothing but a proposition/statement made that we prove whether it is True/False using hypothesis testing.

__Confidence interval:__ is the range of values you expect your estimate to fall between.

__significance value:__ also known as critical value depends up on the domain. By default it is 0.05

__P value:__ is the test value we get after hypothesis testing.

- If distribution is 100% and p_value is 0.05 then (1-0.05 = 0.95) that means confidence interval will be 95%.


__SSR, SSE, SST__
- SSE(Sum of squared error from mean)
- SSR(SUm of squared error from regression line)
- SSt(Sum of total error) SST=SSE+SSR
